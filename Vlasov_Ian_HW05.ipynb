### Задача 1

Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя уровень значимости 0.025 и правостороннюю критическую область, проверить данную гипотезу при условии, что в выборке из 100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 мм.

```python
X_n = 100
T_X = 17.5
D_X = 4
alpha = 0.025
m_H0 = 17
```

Можно утверждать, что случайная величина "диаметр шарика для подшипников" распределена нормально с параметрами $ a = 17 $ (предположение), $ \sigma = 2 $. Также по условию задачи уже выбрана статистика от выборки Х (100 шариков), для которой мы будем проводить статистический тест: среднее от выборки. Для проведения теста необходимо обозначить критечскую область, которая равна 1-0.025 = 0.975 квантилю (правосторонняя область) от функции распределения нашей статистики, которая так же распределена нормально с параметрами  $ a = 17, \sigma = \frac{2}{\sqrt{100}} = 0.2$ Квантиль 0.975 нормального распределния с нашими параметрами найдем по правилу 2 сигм. Вероятность попасть в интервал $(a - 2\sigma, a + 2\sigma)$ равна 0.95, соответственно вероятность попасть в интервал $(a + 2\sigma, \infty)$ равна 0.025. Таким образом критическая область равна:
$$
\Omega_\alpha = (17 + 2\cdot0.2, \infty) = (17.4, \infty)
$$
Так как в результате статистического теста, выборочное среднее получилось равным 17.5, что попадает в критическую область, мы делаем вывод о том, что нулевая гипотеза неверна.

### Задача 2

Известно, что генеральная совокупность распределена нормально с известным средним квадратическим отклонением 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочное среднее равно 80, а объём равен 256.

Требуется найти интервал $
\left(\overline{x}_{в}-\delta ; \overline{x}_{в}+\delta\right)
$,
$\overline{x}_{в} = 80, \sigma = 16, \gamma = 0.95, n = 256$
$$
\delta=\frac{t_{\gamma} \sigma}{\sqrt{n}}, где $$
$$
2 \Phi\left(t_{\gamma}\right)=\gamma = 0.95
$$
$$ t_{\gamma} = 1.96 $$
$$
\delta=\frac{1.96\cdot 16}{\sqrt{256}} = 1.96
$$
Тогда доверительный интервал будет равен:
$$
(80-1.96, 80+1.96) = (78.04, 81.96) $$

### Задача 3

Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:

202, 203, 199, 197, 195, 201, 200, 204, 194, 190

Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что уровень значимости равен 1%?

Так как дисперсия неизвестна, воспользуемся распределением Стьюдетна с параметром $n-1 = 9$. Критическая область для такого распределения и уровнем значимости 1% равна:
$$ (-\infty, t_{\alpha/2, 9}) \cup (t_{1-\alpha/2, 9}, +\infty) = (-\infty, -3.2498) \cup (3.2498, \infty) $$

```python
X = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
print(X.mean())
print(X.std(ddof=1))
```

```python
a = X.std(ddof=1) / np.sqrt(10)
T = (X.mean() - 200) /  a
print(T)
```

Значение не попало в критическую область, значит, утверждение продавца можно считать верным.
