## Непрерывные случайные величины. Функция распределения и плотность распределения вероятностей. Равномерное и нормальное распределение. Центральная предельная теорема

### Задача 1

О непрерывной равномерно распределенной случайной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

Для равномерно распределенной случайной величины известно, что ее дисперсия равна 
$$
D(B)=\frac{(b-a)^{2}}{12}
$$

Подставим, значения:
$$
0.2=\frac{(b-0.5)^{2}}{12}
$$
$$
2.40=b^2 - b + 0.25
$$
$$
b^2 - b - 2.15 = 0
$$
Решив квадратное уравнение, получаем два корня:
$$ b_1 = -1.04919 $$
$$ b_2 = 2.04919 $$

Принимая во внимание, что b - правая граница, то есть $ b > a $ отбрасываем один из корней и получается, что $$ b = 2.04919 $$

### Задача 2

На сколько сигм (средних квадратических отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой $ M(X) = 178 $ и $ D(X) = 25 $?

Пусть рост - это случайная величина. Рост в популяции у конкретного человека, зависит от множества слабо зависимых друг от друга параметров. Считаем, что значения роста распределены нормально. При этом для нормального распределения дисперсия равна  $\sigma $. Тогда 
$$
190 - 178 = 12
$$
$$
\frac{12}{25} = 0.48
$$
Рост человека, равный 190 см, отклонен от математического ожидания $M(X) = 178 $ на $0.48\sigma$

### Задача 3

Рост взрослого населения города X имеет нормальное распределение, причем, средний рост равен 174 см, а среднее квадратическое отклонение равно 8 см. Без использования сторонних библиотек (numpy, scipy, pandas и пр.) посчитайте, какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
1. больше 182 см?
2. больше 190 см?
3. от 166 см до 190 см?
4. от 166 см до 182 см?
5. от 158 см до 190 см?
6. не выше 150 см или не ниже 190 см?
7. не выше 150 см или не ниже 198 см?
8. ниже 166 см?

```python
Image(url= "https://wiki.loginom.ru/images/3-sigma.svg")
```

Подсчет вероятностей в данном задании сводится к верному подсчету интервалов, определенных на картинке выше, где
$$ -3\sigma = 150 $$
$$ -2\sigma = 158 $$
$$ -1\sigma = 166 $$
$$ X_{cp} = 174 $$
$$ 1\sigma = 182 $$
$$ 2\sigma = 190 $$
$$ 3\sigma = 198 $$

1. $ 182 - 174 = 8$ - отклонение от математического ожидания, равно $\sigma$. Известно, что в интервал $1\sigma$ попадает $68.26\%$ выборки. Тогда случайным образом выбранный человек будет иметь рост больше 182 см с вероятностью
$$
P(X > 182) = \frac{100-68.26}{2} = 15.87\%
$$

2. Аналогичное рассуждения только для $2\sigma$. Вероятность будет равна
$$
P(X > 190) = \frac{100-95.44}{2} = 2.28\%
$$

3. В интервал от 166 до 190 см для данного рапределения попадает интервал $-\sigma - +2\sigma$. Вероятность
$$
P(166 > X > 190) = \frac{95.44}{2}+{68.26}{2} = 81.85\%
$$

4. интервал 166 - 182 см равен интервалу $\pm \sigma$, вероятность
$$
P(166 > X > 182) = 68.26\%
$$

5. интервал 158 - 190 см равен интервалу $\pm 2\sigma$, вероятность
$$
P(158 > X > 190) = 95.44\%
$$

6. $$ P(X < 160, X > 190) = \frac{99.72-95.44}{2} + (100-99.72) = 2.42\% $$


7. $$P(X < 160, X > 198) = 100 - 99.72 = 0.28\% $$

8. $$P(X < 166) = \frac{100 - 68.26}{2}=15.87\% $$
